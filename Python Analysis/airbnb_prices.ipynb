# <span style="color:#01686f; font-size:0.75em;">COMBINE ALL THE DIFFERENT DATASETS INTO A SINGLE ONE</span>

```python
def combine(csv_1,col_1,csv_2,col_2,city):       # Combine the 2 datasets into a single dataset.
    csv_1['week time'] = col_1
    csv_2['week time'] = col_2
    csv_1.drop(columns = ['Unnamed: 0'],inplace=True)
    csv_2.drop(columns = ['Unnamed: 0'],inplace=True)
    merged = pd.concat([csv_1, csv_2])
    merged['city'] = city
    return merged
```

<ul>
    <li style="color:#52796f; font-size:0.85em;">
    <h3 >In the above funtion, we combine the 'weekdays' and 'weekend' datasets of the individual datasets into 1
        dataset for a particular city</h3></li>
    <li style="color:#52796f; font-size:0.85em;">
    <h3 >We also add the name of the City and put it into a new column 'city', since we will be combining all these cities datasets and thus would need to differentiate the data of cities in some way for analysis and insights</h3></li>
 <li style="color:#52796f; font-size:0.85em;">
    <h3 >We also remove the unnamed:0 feature since it is the index number of the records, thus is not very useful</h3></li>
</ul>

```python
amsterdam = combine(amsterdam_weekdays,'weekdays',amsterdam_weekends,'weekends','amsterdam')
athens = combine(athens_weekdays,'weekdays',athens_weekends,'weekends','athens')
barcelona = combine(barcelona_weekdays,'weekdays',barcelona_weekends,'weekends','barcelona')
berlin = combine(berlin_weekdays,'weekdays',berlin_weekends,'weekends','berlin')
budapest = combine(budapest_weekdays,'weekdays',budapest_weekends,'weekends','budapest')
lisbon = combine(lisbon_weekdays,'weekdays',lisbon_weekends,'weekends','lisbon')
london = combine(london_weekdays,'weekdays',london_weekends,'weekends','london')
paris = combine(paris_weekdays,'weekdays',paris_weekends,'weekends','paris')
rome = combine(rome_weekdays,'weekdays',rome_weekends,'weekends','rome')
vienna = combine(vienna_weekdays,'weekdays',vienna_weekends,'weekends','vienna')
```

<ul>

    <li style="color:#52796f; font-size:0.85em;"><h3>For each city, we collect listing URLs from the search results, open each listing page, and extract the live listing title, nightly price, rating, reviews count, Superhost flag, Guest Favorite flag, listing URL, and scrape timestamp.</h3></li>
    <li style="color:#52796f; font-size:0.85em;"><h3>The scraped data is stored in a separate DataFrame and exported to its own CSV. It is not merged directly with the old Kaggle records.</h3></li>

```python
# Selenium is used only for the new live Airbnb data.
# If Selenium is not installed in your environment, run:
# !pip install selenium webdriver-manager

from datetime import datetime
from urllib.parse import quote_plus
import re
import time

try:
    from selenium import webdriver
    from selenium.webdriver.common.by import By
    from selenium.webdriver.chrome.options import Options
    from selenium.webdriver.chrome.service import Service
    from selenium.webdriver.support.ui import WebDriverWait
    from selenium.webdriver.support import expected_conditions as EC
    from webdriver_manager.chrome import ChromeDriverManager
except ImportError:
    webdriver = None
    print("Selenium/webdriver-manager are not installed. Install them before running the scraping cells.")
```

```python
# Use the original data only to decide which cities will be searched on Airbnb.
scrape_target_cities = sorted(
    europe_data[
        (europe_data['room_type'].str.lower() == 'entire home/apt') &
        (europe_data['bedrooms'] == 2)
    ]['city'].dropna().unique()
)

MAX_LISTINGS_PER_CITY = 50
SCRAPED_AIRBNB_CSV = 'Scraped_Airbnb_Listings.csv'

scrape_target_cities
```

```python
def build_airbnb_search_url(city, offset=0):
    city_query = quote_plus(f"{city.title()}, Europe")
    return (
        f"https://www.airbnb.com/s/{city_query}/homes"
        "?room_types%5B%5D=Entire%20home%2Fapt"
        "&min_bedrooms=2"
        f"&items_offset={offset}"
    )


def start_airbnb_driver(headless=False):
    if webdriver is None:
        raise ImportError("Install selenium and webdriver-manager before starting the scraper.")

    chrome_options = Options()
    if headless:
        chrome_options.add_argument("--headless=new")
    chrome_options.add_argument("--window-size=1440,1200")
    chrome_options.add_argument("--disable-blink-features=AutomationControlled")
    chrome_options.add_argument("--disable-notifications")
    chrome_options.add_argument("--lang=en-US")

    service = Service(ChromeDriverManager().install())
    return webdriver.Chrome(service=service, options=chrome_options)


def close_cookie_banner(driver):
    button_xpaths = [
        "//button[contains(translate(., 'ABCDEFGHIJKLMNOPQRSTUVWXYZ', 'abcdefghijklmnopqrstuvwxyz'), 'accept')]",
        "//button[contains(translate(., 'ABCDEFGHIJKLMNOPQRSTUVWXYZ', 'abcdefghijklmnopqrstuvwxyz'), 'ok')]",
        "//button[contains(translate(., 'ABCDEFGHIJKLMNOPQRSTUVWXYZ', 'abcdefghijklmnopqrstuvwxyz'), 'got it')]",
    ]
    for xpath in button_xpaths:
        try:
            WebDriverWait(driver, 3).until(EC.element_to_be_clickable((By.XPATH, xpath))).click()
            break
        except Exception:
            pass


def collect_listing_links(driver, city, max_links=50):
    links = []
    seen = set()
    offset = 0

    while len(links) < max_links:
        driver.get(build_airbnb_search_url(city, offset))
        close_cookie_banner(driver)
        time.sleep(5)

        anchors = driver.find_elements(By.XPATH, "//a[contains(@href, '/rooms/')]")
        for anchor in anchors:
            href = anchor.get_attribute("href")
            if not href:
                continue

            clean_url = href.split("?")[0]
            if "/rooms/" in clean_url and clean_url not in seen:
                links.append(clean_url)
                seen.add(clean_url)
                if len(links) >= max_links:
                    break

        if len(anchors) == 0 or len(links) >= max_links:
            break

        offset += 18
        time.sleep(2)

    return links[:max_links]
```

```python
def first_text(driver, selectors):
    for by, selector in selectors:
        try:
            value = WebDriverWait(driver, 4).until(
                EC.presence_of_element_located((by, selector))
            ).text.strip()
            if value:
                return value
        except Exception:
            pass
    return None


def extract_price(text):
    if not text:
        return None
    match = re.search(r"([\$\u20ac\u00a3]|EUR|GBP)\s?([0-9][0-9,]*)", text)
    if match:
        return float(match.group(2).replace(",", ""))
    return None


def extract_rating(text):
    if not text:
        return None
    patterns = [
        r"(\d\.\d+)\s*(?:out of 5|rating|stars?)",
        r"(\d\.\d+)\s*Â·",
    ]
    for pattern in patterns:
        match = re.search(pattern, text, flags=re.IGNORECASE)
        if match:
            return float(match.group(1))
    return None


def extract_reviews_count(text):
    if not text:
        return None
    match = re.search(r"([0-9][0-9,]*)\s+reviews?", text, flags=re.IGNORECASE)
    if match:
        return int(match.group(1).replace(",", ""))
    return None


def scrape_listing_details(driver, url, city):
    driver.get(url)
    time.sleep(4)
    close_cookie_banner(driver)

    page_text = driver.find_element(By.TAG_NAME, "body").text
    title = first_text(driver, [
        (By.TAG_NAME, "h1"),
        (By.CSS_SELECTOR, "[data-testid='listing-page-title']"),
    ])
    price_text = first_text(driver, [
        (By.CSS_SELECTOR, "[data-testid='price-availability-row']"),
        (By.XPATH, "//*[contains(text(), '$') or contains(text(), 'EUR') or contains(text(), 'GBP')]"),
    ])

    return {
        "city": city,
        "listing_title": title,
        "nightly_price": extract_price(price_text) or extract_price(page_text),
        "rating": extract_rating(page_text),
        "reviews_count": extract_reviews_count(page_text),
        "host_is_superhost": bool(re.search(r"superhost", page_text, flags=re.IGNORECASE)),
        "guest_favorite": bool(re.search(r"guest favorite", page_text, flags=re.IGNORECASE)),
        "room_type_filter": "Entire home/apt",
        "minimum_bedrooms_filter": 2,
        "listing_url": url,
        "scraped_at": datetime.utcnow().strftime("%Y-%m-%d %H:%M:%S"),
    }
```

```python
# Run this cell when you are ready to collect fresh Airbnb data.
# Airbnb pages can change and may show CAPTCHA/rate limits, so keeping the browser visible is useful.

scraped_records = []

if webdriver is not None:
    driver = start_airbnb_driver(headless=False)
    try:
        for city in scrape_target_cities:
            print(f"Scraping {city}...")
            listing_links = collect_listing_links(driver, city, max_links=MAX_LISTINGS_PER_CITY)
            print(f"Found {len(listing_links)} listing links for {city}.")

            for listing_url in listing_links:
                try:
                    scraped_records.append(scrape_listing_details(driver, listing_url, city))
                except Exception as exc:
                    print(f"Skipped {listing_url}: {exc}")
                time.sleep(1.5)
    finally:
        driver.quit()

scraped_airbnb_data = pd.DataFrame(scraped_records)
scraped_airbnb_data.to_csv(SCRAPED_AIRBNB_CSV, index=False)
scraped_airbnb_data.head()
```

# **Scraped Airbnb Data Warehouse Tables**

<ul>
<li style="color:#52796f; font-size:0.85em;"><h3>The new scraped Airbnb features are modeled in their own fact table and dimensions.</h3></li>
<li style="color:#52796f; font-size:0.85em;"><h3>City can be shared with the original warehouse, but the live listing snapshot remains separate from the historical Kaggle fact table.</h3></li>
</ul>

```python
# Dimension source
dim_source = pd.DataFrame({
    'source_id': [1, 2],
    'source_name': ['Original European Airbnb CSVs', 'Live Airbnb Selenium Scrape'],
    'source_type': ['historical_dataset', 'live_web_scrape']
})

dim_source
```

```python
# Extend city dimension with any city found in scraped data.
all_cities_for_warehouse = sorted(
    set(dim_city['city'].dropna()) | set(scraped_airbnb_data['city'].dropna())
)

dim_city_extended = pd.DataFrame({'city': all_cities_for_warehouse})
dim_city_extended['city_id'] = dim_city_extended.index + 1

dim_city_extended
```

```python
# Dimension for scraped listing identity and search filters
dim_scraped_listing = (
    scraped_airbnb_data[
        [
            'listing_url',
            'listing_title',
            'city',
            'room_type_filter',
            'minimum_bedrooms_filter'
        ]
    ]
    .drop_duplicates()
    .reset_index(drop=True)
)

dim_scraped_listing['scraped_listing_id'] = dim_scraped_listing.index + 1
dim_scraped_listing = dim_scraped_listing[
    [
        'scraped_listing_id',
        'listing_url',
        'listing_title',
        'city',
        'room_type_filter',
        'minimum_bedrooms_filter'
    ]
]

dim_scraped_listing
```

```python
# Dimension for live scraped host/quality flags
dim_scraped_listing_flags = (
    scraped_airbnb_data[
        [
            'host_is_superhost',
            'guest_favorite'
        ]
    ]
    .drop_duplicates()
    .reset_index(drop=True)
)

dim_scraped_listing_flags['scraped_flags_id'] = dim_scraped_listing_flags.index + 1
dim_scraped_listing_flags
```

```python
# Fact table for the live scraped listing snapshot
fact_scraped_airbnb = (
    scraped_airbnb_data
    .merge(
        dim_scraped_listing[['scraped_listing_id', 'listing_url']],
        on='listing_url',
        how='left'
    )
    .merge(
        dim_city_extended,
        on='city',
        how='left'
    )
    .merge(
        dim_scraped_listing_flags,
        on=['host_is_superhost', 'guest_favorite'],
        how='left'
    )
)

fact_scraped_airbnb['source_id'] = 2
fact_scraped_airbnb['scrape_record_id'] = range(1, len(fact_scraped_airbnb) + 1)

fact_scraped_airbnb = fact_scraped_airbnb[
    [
        'scrape_record_id',
        'source_id',
        'city_id',
        'scraped_listing_id',
        'scraped_flags_id',
        'nightly_price',
        'rating',
        'reviews_count',
        'scraped_at'
    ]
]

fact_scraped_airbnb
```

```python
# Export the new scraped warehouse tables as separate CSV files.
scraped_airbnb_data.to_csv('Scraped_Airbnb_Listings.csv', index=False)
dim_source.to_csv('Dim_Source.csv', index=False)
dim_city_extended.to_csv('Dim_City_Extended.csv', index=False)
dim_scraped_listing.to_csv('Dim_Scraped_Listing.csv', index=False)
dim_scraped_listing_flags.to_csv('Dim_Scraped_Listing_Flags.csv', index=False)
fact_scraped_airbnb.to_csv('Fact_Scraped_Airbnb.csv', index=False)

pd.DataFrame({
    'table_name': [
        'Scraped_Airbnb_Listings',
        'Dim_Source',
        'Dim_City_Extended',
        'Dim_Scraped_Listing',
        'Dim_Scraped_Listing_Flags',
        'Fact_Scraped_Airbnb'
    ],
    'rows': [
        len(scraped_airbnb_data),
        len(dim_source),
        len(dim_city_extended),
        len(dim_scraped_listing),
        len(dim_scraped_listing_flags),
        len(fact_scraped_airbnb)
    ]
})
```
